##### CREATE TABLES

```python
# Create tables in the database

import psycopg2
connection = psycopg2.connect(
    host = 'localhost',
    dbname = 'postgres',
    user = 'postgres',
    port = 5432,
    password = 'Ahsan'
)
cursor = connection.cursor()

with connection.cursor() as cursor:
    cursor.execute('''CREATE TABLE IF NOT EXISTS Author (
           id INT PRIMARY KEY,
           name VARCHAR(50),
           birthdate DATE,
           nationality VARCHAR(50))
           ''')
    
    cursor.execute('''CREATE TABLE IF NOT EXISTS Books (
           id INT PRIMARY KEY,
           title VARCHAR(50),
           author_id INT,
           genre VARCHAR(50),
           publication_year INT,
           available_copies INT DEFAULT 0)
           ''')
    
    cursor.execute('''CREATE TABLE IF NOT EXISTS Members (
           id INT PRIMARY KEY,
           name VARCHAR(50) NOT NULL,
           membership_date DATE NOT NULL,
           email VARCHAR(50))
           ''')
    
    connection.commit()

print("Tables created successfully.")
```

## QUERIES:

```python
import psycopg2
connection = psycopg2.connect(
    host = 'localhost',
    dbname = 'postgres',
    user = 'postgres',
    port = 5432,
    password = 'Ahsan'
)
cursor = connection.cursor()

#SELECT ALL 

cursor.execute('SELECT * from Author')
authors = cursor.fetchall()
for author in authors:
    print(author)

#SELECT NAME,NATIONALITY OF AUTHOR

cursor.execute('SELECT name,nationality from Author')
authors = cursor.fetchall()
for author in authors:
    print(author)

# AVAILABLE COPIES OF BOOKS

cursor.execute('SELECT * FROM Books where available_copies > 0')
books = cursor.fetchall()
for book in books:
    print(book)

# Count Total Books:

cursor.execute('Select Count(*) from Books')
books = cursor.fetchall()
for book in books:
    print(book)

# Count Books by Genre:

cursor.execute("Select Count(*) from Books where genre = 'Interesting' ")
books = cursor.fetchall()
for book in books:
    print(book)

# Inner Join (Books with Authors)

cursor.execute('SELECT b.title, a.name AS author FROM Books b JOIN Authors a ON b.author_id = a.id;')
result = cursor.fetchall()
for rows in result:
    print(rows)

# Left Join (All Authors and Their Books)

cursor.execute('SELECT a.name, b.title FROM Authors a LEFT JOIN Books b ON a.id = b.author_id;')
result = cursor.fetchall()
for rows in result:
    print(rows)

#Right Join() Books and Their Authors)

cursor.execute("""SELECT b.title, a.name FROM Books b RIGHT JOIN Authors a ON b.author_id = a.id;""")
result = cursor.fetchall()
for rows in result:
    print(rows)


```
